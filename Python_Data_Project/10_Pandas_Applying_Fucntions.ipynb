```python
import pandas as pd
import matplotlib.pyplot as plt
import ast


url = "https://huggingface.co/datasets/lukebarousse/data_jobs/resolve/main/data_jobs.csv"
df = pd.read_csv(url, dtype={'job_skills': str})
df = df.iloc[1:]
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python

def safe_eval(x):
    try:
        # This catches real NaNs and strings like 'nan'
        if isinstance(x, str) and x != 'nan':
            return ast.literal_eval(x)
    except (ValueError, SyntaxError):
        pass
    return None
```

```python
df['job_skills'] = df['job_skills'].astype(str)
```

```python
df['job_skills'] = df['job_skills'].apply(safe_eval)
```
